In Heston model characterstic function looks like
$$\phi_H (u, T; v(t_0)) = 
\exp \left[ \left( iur\tau + \dfrac{v(t_0)}{\gamma^2} \left( \dfrac{1 - e^{-D_1 \tau}}{1 - g e^{-D_1 \tau}} \right) \right) \left( \kappa - i \rho \gamma u - D_1 \right) \right] 
\times 
\exp \left[  \dfrac{\kappa \bar v}{\gamma^2} \left(\tau(\kappa - i \rho \gamma u - D_1) - 2\log \left( \dfrac{1 - g e^{-D_1 \tau}}{1 - g}\right) \right)\right],$$

$$\tau = T - t_0,$$

$$D_1 = \sqrt{(\kappa - i \rho \gamma u)^2 + (u^2 + iu) \gamma^2},
g = \dfrac{\kappa - i \rho \gamma u - D_1}{\kappa - i \rho \gamma u + D_1},$$

$$v(t_0) \text{ - variance at } t_0.$$

**Note**: in out setup we have $t_0 = 0$ and $\tau = T$.

```python
heston_parameters = HestonParameters(kappa = 1.3125, gamma = 0.7125, rho = -0.3937, vbar = 0.0641, v0 = 0.1181)
state = MarketState(stock_price = 1, interest_rate = 0.01, dividends_rate = 0)


strikes = np.linspace(50, 150, 50)
times = np.linspace(0, 2, 50)
calls = CallStockOption(strike_price=strikes.reshape(1, -1),
                        expiration_time=times.reshape(-1, 1))

char_func = characteristic_function_Heston_model(state=state, 
                                                 heston_parameters=heston_parameters)

u = 1  #initialize arbitrary argument
maturities = 1
assert isinstance(char_func(u, maturities), complex), "Mismatch in output types"

u = np.array([1,2,3])  #initialize arbitrary argument
maturities = 1
assert char_func(u, maturities).shape == u.shape, "Mismatch in shapes"

u = 1  #initialize arbitrary argument
maturities = calls.expiration_time
assert char_func(u, maturities).shape == maturities.shape, "Mismatch in shapes"

u = np.array([1,2,3]) #initialize arbitrary argument
maturities = calls.expiration_time
assert char_func(u, maturities).shape == (maturities.shape[0], u.shape[0]), "Mismatch in shapes"
```

Cosine series coefficfients are 
$$\chi_k(c, d) := \dfrac{1}{1 + (\dfrac{k \pi}{b - a})^2} \left[ \cos(k \pi \dfrac{d - a}{b - a})e^{d} - \cos(k \pi \dfrac{c - a}{b - a})e^{c} + 
\dfrac{k \pi}{b - a} \sin (k \pi \dfrac{d - a}{b - a}) e^{d} - \dfrac{k \pi}{b - a} \sin(k \pi \dfrac{c-a}{b-a}) e^{c} \right]$$

$$\psi_k(c, d) := 
\begin{cases}
 \left[ \sin(k \pi \dfrac{d - a}{b - a}) - \sin(k \pi \dfrac{c - a}{b - a}) \right] \dfrac{b - a}{k \pi}, k \neq 0 \\
 d - c, k = 0
\end{cases}
$$
$$
[c,d] \subset [a,b]
$$

```python
a, b, c, d = 1, 2, 3, 4
k = np.linspace(0, 2, 50)
cos_ser_coefficients = cosine_series_coefficients(a, b, c, d, k)

assert cos_ser_coefficients["chi"].shape == k.shape, "Chi shape mismatch"
assert cos_ser_coefficients["psi"].shape == k.shape, "Psi shape mismatch"
```

Coefficients for option prices in Fourier expansion are:

$$H^{call}_k = \dfrac{2}{b - a} (\chi_k(0, b) - \psi_k(0, b))$$
$$H^{put}_k = \dfrac{2}{b - a} (-\chi_k(a, 0) - \psi_k(a, 0))$$

```python
a, b = 1, 2
k = np.linspace(0, 2, 50)
opt_coefficients = options_coefficients(True, a, b, k)

assert opt_coefficients.shape == k.shape, "Coefficients shape mismatch"
```

In Heston model one can derive a very good approximation of option price using COS formula

$$V(t_0, x) = K e^{-r \tau} Re \left\{ \sum_{k=0}^{N-1} {'} \phi_H \left( \dfrac{k \pi}{b - a}, T; v(t_0) \right) U_k \cdot \exp \left( ik\pi \dfrac{X(t_0) - a}{b - a} \right) \right\}$$

where
$$X(t_0) = \log(\dfrac{S(t_0)}{K}),$$


Deltas are

$$\Delta \simeq \dfrac{1}{S(t_0)} e^{-r \tau} Re \left\{ \sum_{k=0}^{N-1} {'} \phi_H \left( \dfrac{k \pi}{b - a}, T; v(t_0) \right) U_k \cdot \exp \left( -ik\pi \dfrac{X(t_0) - a}{b - a} \right) \dfrac{ik\pi}{b - a} \right\}$$


```python
#initialize market parameters
heston_parameters = HestonParameters(kappa = 1.3125, gamma = 0.7125, rho = -0.3937, vbar = 0.0641, v0 = 0.1181)
state = MarketState(stock_price = 100, interest_rate = 0.01, dividends_rate = 0)

#initialize characteristic function
characteristic_function = characteristic_function_Heston_model(state, heston_parameters)

#initialize grid of options
strikes = np.linspace(50, 150, 50)
times = np.linspace(0.1, 2, 50)

calls = CallStockOption(strike_price=strikes.reshape(1, -1),
                        expiration_time=times.reshape(-1, 1))

puts = PutStockOption(strike_price=strikes.reshape(1, -1),
                      expiration_time=times.reshape(-1, 1))

call_prices_coss, call_deltas = COS_method(char_function = characteristic_function_Heston_model(state, heston_parameters),
                                           option = calls, state = state,
                                           L = 10, N = 10000, greek_flag=True)

put_prices_coss, put_deltas = COS_method(char_function = characteristic_function_Heston_model(state, heston_parameters),
                                         option = puts, state = state,
                                         L = 10, N = 10000, greek_flag=True)
```

```python
fig = plt.figure(figsize=(15, 35))

x, y = np.meshgrid(strikes, times)

ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(x, y, call_prices_coss.squeeze())
ax1.set_title('Call option price with COS')
ax1.set_xlabel('Strike price, $')
ax1.set_ylabel('Time to expiration, years')
ax1.set_zlabel('Option price, $')

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(x, y, put_prices_coss.squeeze())
ax2.set_title('Put option price with COS')
ax2.set_xlabel('Strike price, $')
ax2.set_ylabel('Time to expiration, years')
ax2.set_zlabel('Option price, $')
```
